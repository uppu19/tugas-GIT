```python
data_panen = {
    'lokasi1': {
        'nama_lokasi' : 'Kebun A',
        'hasil_panen' : {
            'padi' : 1200,
            'jagung' : 800,
            'kedelai' : 500
        }
    },
    'lokasi2': {
        'nama_lokasi' : 'Kebun B',
        'hasil_panen' : {
            'padi' : 1500,
            'jagung' : 900,
            'kedelai' : 450
        }
    },
    'lokasi3': {
        'nama_lokasi' : 'Kebun C',
        'hasil_panen' : {
            'padi' : 1200,
            'jagung' : 700,
            'kedelai' : 600
        }
    },
    'lokasi4': {
        'nama_lokasi' : 'Kebun D',
        'hasil_panen' : {
            'padi' : 1300,
            'jagung' : 850,
            'kedelai' : 550
        }
    },
    'lokasi5': {
        'nama_lokasi' : 'Kebun E',
        'hasil_panen' : {
            'padi' : 1400,
            'jagung' : 950,
            'kedelai' : 480
        }
    }
}

# mencetak data hasil panen
for lokasi in data_panen.values():
    print(f"Data Hasil Panen di {lokasi['nama_lokasi']}:")
    for hasil in lokasi['hasil_panen'].items():
        print(f" {hasil} kg")
    
#mencetak  hasil jaugung di lokasi 2
lokasi2 = data_panen['lokasi2']
jumlah_hjagung_lokasi2 = lokasi2['hasil_panen']['jagung']
print(f"Jumlah hasil panen jagung lokasi 2: {jumlah_hjagung_lokasi2} kg")

#mencetak nama lokasi 3
lokasi3 = data_panen['lokasi3']
nama_lokasi3 = lokasi3['nama_lokasi']
print(f"Nama lokasi dari lokasi3 adalah {nama_lokasi3}")

#memasukan ke variabel yang berbeda
jumlah_padi = []
jumlah_kedelai = []

for lokasi in data_panen.values():
    jumlah_padi.append(lokasi['hasil_panen']['padi'])
    jumlah_kedelai.append(lokasi['hasil_panen']['kedelai'])

print("Jumlah hasil panen padi di setiap lokasi:", jumlah_padi)
print("Jumlah hasil panen kedelai di setiap lokasi:", jumlah_kedelai)

```
